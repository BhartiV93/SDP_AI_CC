### LAB 10 - Implementing Seaborn Concepts in Python

**Problem 1:** How to add the title to the plot using the Matplotlib.

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
  
plt.show()
```

**Problem 2:** How to set the xlim and ylim

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# setting the x limit of the plot
plt.xlim(5)
  
plt.show()
```

**Problem 3:** How to use the dark theme

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# changing the theme to dark
sns.set_style("dark")
plt.show()
```

**Problem 4:** How to Change the figure Size

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# changing the figure size
plt.figure(figsize = (2, 4))
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
  
plt.show()
```

**Problem 5:** How to Diverge the  Color Palette

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt
  
# current colot palette
palette = sns.color_palette('PiYG', 11)
  
# diverging color palette
sns.palplot(palette)
  
plt.show()
```

**Problem 6:** How to create a relational plots

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
  
# loading dataset
data = sns.load_dataset("iris")
  
# creating the relplot
sns.relplot(x='sepal_width', y='species', data=data)
  
plt.show()
```
