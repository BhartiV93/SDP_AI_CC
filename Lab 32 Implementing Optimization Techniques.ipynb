Generally, optimization in a deep neural network are:

Let's use the optimization techniques on the model developed in previous Lab exercise.

So, let's develop a DL model. Dataset, used here, can be downloaded from this link: https://www.kaggle.com/datasets/karakaggle/kaggle-cat-vs-dog-dataset

```python
# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense


from keras import regularizers
from keras.layers.core import Dropout
from keras.callbacks import EarlyStopping
```

```python
#Initialising the CNN
classifier = Sequential()

# Step 1 - Convolution
classifier.add(Convolution2D(32, 3, 3, input_shape = (64,64, 3), activation = 'relu'))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer
classifier.add(Convolution2D(32, 3, 3, activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full connection
classifier.add(Dense(128, activation = 'relu'))
classifier.add(Dense(1, activation = 'sigmoid'))

```

```python
from keras.preprocessing.image import ImageDataGenerator

# Data Augmentation

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory('data/train',
                                                 target_size = (64,64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

test_set = test_datagen.flow_from_directory('data/test',
                                            target_size = (64,64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
# Compiling the CNN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

model_variation1= classifier.fit(training_set,# the training set
                         epochs = 100,# number of epochs 
                         validation_data = test_set# the test set
                         )
```

Final validation accuracy is around 0.7

In the above implemtation, we will add dropout layer, regularizer and early stopping.

Remember, dropout layer, regularizer and early stopping not always guaranty better results. In general, they work best in large networks.

```python
#Initialising the CNN
classifier1 = Sequential()

# Step 1 - Convolution
classifier1.add(Convolution2D(32, 3, 3, input_shape = (64,64, 3), activation = 'relu'))

# Step 2 - Pooling
classifier1.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer
classifier1.add(Convolution2D(32, 3, 3, activation = 'relu',kernel_regularizer=regularizers.l1(0.0001)))
classifier1.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
classifier1.add(Flatten())

# Step 4 - Full connection
classifier1.add(Dense(128, activation = 'relu'))
classifier1.add(Dropout(0.25))
classifier1.add(Dense(1, activation = 'sigmoid'))

```

```python
classifier1.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model_variation2= classifier1.fit(training_set, epochs=100, validation_data=test_set, callbacks = [EarlyStopping(monitor='val_accuracy', patience=2)])
```
